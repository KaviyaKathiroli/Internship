```python
response=requests.get("https://www.flipkart.com/sunglasses/pr?sid=26x&q=sunglasses&otracker=categorytree/")
soup=BeautifulSoup(response.text,'html.parser')
print(soup.prettify())
```

```python
alldivs=soup.find('div')
alldivs.prettify()
```

To get all classes in the soup

```python
class_list = set()
tags = {tag.name for tag in soup.find_all()}
for tag in tags:
    #print(tag)
    for i in soup.find_all(tag):
        #print(i)
        if(i.has_attr("class")):
            #print(i)
            class_list.add(" ".join( i['class']))
type(class_list)           
class_list
```

```python
for i in class_list:
    list_tags = soup.find_all(class_=i)
    print(list_tags)
```
